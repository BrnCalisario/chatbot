```python
import pandas as pd
import numpy as np
```

```python
# Load Model
root_df = pd.read_csv("data.csv")
root_df
```

```python
# Cleaning stop words

clean_df = root_df.copy()

stop_words = pd.read_csv("stop-words.txt").squeeze()

for word in stop_words:
    clean_df["Message"] = clean_df["Message"].apply(lambda x : x.replace(" " + word + " ", ' '))

clean_df
```

```python
# SkLearn Imports
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import make_pipeline
```

```python
# Vectorizing X

vec = CountVectorizer(lowercase=True)
X = vec.fit_transform(raw_documents=clean_df["Message"]).toarray()
```

```python
# Enconding Y

le = LabelEncoder()
Y = le.fit_transform(clean_df["Intent"])

Y
```

```python
model = make_pipeline(CountVectorizer(lowercase=True), MultinomialNB())

model.fit(clean_df["Message"], Y)
```

```python
# Tests

new_message = [input()]

predict = model.predict(new_message)

print("Predict: " + le.inverse_transform(predict)[0])
print(model.predict_proba(new_message))
```
